the cell below imports the libraries needed to perform pi, sin, cos, and linspace

```python
import math as m
import numpy as np
```

the function bellow returns sin(x)

```python
def return_sin(x):
    return m.sin(x)
```

the function below returns cos(x)

```python
def return_cos(x):
    return m.cos(x)
```

the below function will return a set of the arrays with 1000 entries that contain the values of x, sin(x), and cos(x)

```python
def table(start, end, entry_number):
    x_entries = np.linspace(0, m.pi * 2, entry_number)
    sin_entries = np.linspace(return_sin(0), return_sin(m.pi * 2), entry_number)
    cos_entries = np.linspace(return_cos(0), return_cos(m.pi * 2), entry_number)
    return (x_entries, sin_entries, cos_entries)
```

the main function below prints the first ten entries of the values of x, sin(x), and cos(x)

```python
def main():
    x_entries, sin_entries, cos_entries = table(0, m.pi * 2, 1000)
    for i in range(0, 10):
        print(f"{x_entries[i]}\t{sin_entries[i]}\t{cos_entries[i]}")
```

the cell below makes sure that the main function runs

```python
if __name__=="__main__":
    main()
```
