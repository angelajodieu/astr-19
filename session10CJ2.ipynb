```python
#import matplotlib and other libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#favorite distribution: np.random.power
#I like the formula ax^(a-1)

#pull 1000 random numbers
samples = 1000
x = np.random.power(5, 1000)

#histogram random numbers into 100 bins
bins = np.linspace(0, 1, 100)

#plot and label histogram
plt.hist(x, bins=bins, alpha=0.5, edgecolor="black")
plt.ylabel("N per bin")
plt.xlabel("x")

#save figure as PDF
plt.savefig("session10prompt.pdf")
```
