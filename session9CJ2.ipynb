```python
#import matplotlib and other libraries
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
#pull 1000 random numbers distributed uniformly between [0,1]
n = 1000 #number of random data points to generate
x = np.random.randn(n)#generate random data

#histogram the random numbers into 100 bins
bins = np.linspace(0, 1, 100)

#plot the histogram and label
plt.hist(x, bins=bins, alpha=0.5, edgecolor="black")
plt.ylabel("N per bin")
plt.xlabel("x")

#save figure as a PDF
plt.savefig("session9CJ2.pdf", bbox_inches="tight", facecolor="white")
```
